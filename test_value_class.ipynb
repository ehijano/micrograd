```python
# Example of some complicated function L = tanh( (a*b + c)*f )
a = Value(2.0, label = 'a')
b = Value(-3.0, label = 'b')
c = Value(10.0, label = 'c')

d = a*b + c
d.label = 'd'

f1 = Value(-2.0, label = 'f1')
f2 = Value(7.0, label = 'f2')

g = d * f1 + f2
g.label = 'g'


L  = g.tanh()
L.label = 'L'

gr.draw_dot(L)
```

```python
L.backward()

gr.draw_dot(L)
```
